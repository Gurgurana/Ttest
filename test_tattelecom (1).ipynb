#### Топ -10 городов по числу клиентов

Рассмотрим лидеров по числу клиентов среди всех городов в относительных значениях


```python
city_distr = pd.pivot_table(df, index = ['city'], values = ['id'], aggfunc = 'nunique').reset_index()
city_distr_10 = city_distr.sort_values(by = ['id'], ascending = False).head(10)
city_distr_10['share'] = round(city_distr_10['id']/city_distr['id'].sum(),2)


city_distr_10

```

```python
def pie_city(x, y, title):
    city = city_distr.groupby(x)[y].sum().sort_values(ascending=False)
    city_top = city[:10]
    city_top['other'] = city.sum()-city[:10].sum()
    city_top.plot(kind='pie', autopct='%1.0f%%');
    plt.title(title)
    return plt.show()
```

```python
pie_city('city', 'id', 'Топ-10 городов')
```

Посмотрим на абсолютные значения, которые иллюстрирует столбчатая диаграмма:

```python
city_distr_10.plot(x = 'city', y = 'id', kind = 'bar', figsize=(20,8), grid=True)

plt.title('Распределение клиентов по крупнейшим городам', fontsize= 18)
plt.ylabel('Количество клиентов', fontsize=14)
plt.tick_params(labelsize=14)
plt.xlabel('Наименование населенного пункта', fontsize=14)
plt.xticks(rotation=90)
plt.show();
```

Населенных пунктов в татарстане достаточно много, поэтому в данном пункте мы рассмотрели крупнейшие из них, и выяснили следующее:\
Тройку городов-лидеров возглавляют Казань, Набережные челны, и Нижнекамск.
Число клиентов из населенных пунктов, не вошедших в десятку лидеров, составляет 13% от общего числа.
